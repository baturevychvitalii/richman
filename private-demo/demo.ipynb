## Accounts

```python
# Spending
fun = Account('Fun')
wallet = Account('Wallet', sub_accounts=[fun])

# Savings
safety = Account('Safety')
holiday = Account('Holiday')
style = Account('Style')
learning = Account('Learning')
car_insurance = Account('Car Insurance')
car_maintenance = Account('Car Maintenance')
car = Account('Car', sub_accounts=[car_insurance, car_maintenance])
savings = Account('Savings', sub_accounts=[safety, holiday, style, learning, car])
```
